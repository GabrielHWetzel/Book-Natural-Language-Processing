# Load book

```python
import re
with open ("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

```python
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
```

## Common words that are not articles

```python
pattern = re.compile("[a-zA-Z]+")
result = re.findall(pattern, book.lower())
```

```python
words = {}
for word in result:
    if word in words.keys():
        words[word] = words[word] + 1
    else:
        words[word] = 1
```

```python
words_list = [(value, key) for (key, value) in words.items()]
```

```python
words_n_article = []
for count, word in words_list:
    if word not in english_stopwords:
        words_n_article.append((count, word))
```

```python
sorted(words_n_article, reverse=True)
```

## Sentiment Analysis
What is the most positive and negative sentences

```python
from nltk.sentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.polarity_scores(book)
```

```python
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
```

```python
for nmbr, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(nmbr, scores)
```
